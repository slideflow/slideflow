```python
import pandas as pd
import numpy as np

df = pd.read_csv("/home/prajval/DATA/PROJECTS/TCGA_LUNG/eval/ensemble_evaluations/4/tile_predictions_ensemble.csv")
df.head()
```

```python
slide_bit = df['slide']
slide = list(slide_bit)
```

```python
mod_slide = []
for i in slide:
    mod_slide.append(i[2:-1])
# mod_slide
```

```python
df.insert(loc=1, column='slide_mod', value= mod_slide)
df[["slide_mod","slide"]]
```

```python
df = df.drop(['slide', 'Unnamed: 0'], axis=1)
```

```python
df.head()
```

```python
df = df.rename(columns={"slide_mod": "slide"})
```

```python
df.head()
```

```python
df.to_csv("/home/prajval/DATA/PROJECTS/TCGA_LUNG/eval/ensemble_evaluations/4/tile_predictions.csv")
```

```python
import slideflow as sf

P = sf.Project("/home/prajval/DATA/PROJECTS/TCGA_BRCA/")
labels,unique = P.dataset().labels("histological_type", format = "name")
```

```python
df["histological_type"] = df.slide.map(labels)
df["y_true"] = (df.histological_type == "Infiltrating Lobular Carcinoma").astype(int)
```

```python
sdf = df.loc[df.histological_type.isin(["Infiltrating Ductal Carcinoma", "Infiltrating Lobular Carcinoma"])]
```

```python
sdf
```

```python
from slideflow.stats.metrics import ClassifierMetrics

metrics = ClassifierMetrics(df.y_true.values, df["histological_type-y_pred1"].values)
print(metrics.auroc)
```

```python
df["in_v_out"] = (~df.histological_type.isin(["Infiltrating Ductal Carcinoma", "Infiltrating Lobular Carcinoma"])).astype(int)
```

```python
df
```

## softmax_unc

```python
import numpy as np
df["softmax_unc"] = 0.5 - np.abs(0.5 - df["histological_type-y_pred1"].values)
```

```python
from slideflow.stats.metrics import ClassifierMetrics

metrics = ClassifierMetrics(df.in_v_out.values, df["softmax_unc"].values)
print(metrics.auroc)
```
